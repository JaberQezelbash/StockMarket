* **The LSTM model**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix
from keras.models import Sequential
from keras.layers import Dense, LSTM
import matplotlib.pyplot as plt
import os

# File directories
data_path = r"C:\Users\Jaber\OneDrive - University of Florida\Educational\GitHub\StockMarket\Data\all_stocks_5yr.csv"
results_dir = r"C:\Users\Jaber\OneDrive - University of Florida\Educational\GitHub\StockMarket\Results"
united_csv_path = os.path.join(results_dir, 'all_stocks_predictions_metrics.xlsx')

# Create results directory if it doesn't exist
if not os.path.exists(results_dir):
    os.makedirs(results_dir)

# Load dataset
df = pd.read_csv(data_path)

# Data Preprocessing
def preprocess_data(df, stock_name):
    df_stock = df[df['stock'] == stock_name].copy()
    df_stock.sort_values('date', inplace=True)
    
    # Scale data
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(df_stock[['close']].values)
    
    return df_stock, scaled_data, scaler

# Create the dataset for LSTM
def create_dataset(scaled_data, time_step=60):
    X, y = [], []
    for i in range(len(scaled_data) - time_step):
        X.append(scaled_data[i:i + time_step, 0])
        y.append(scaled_data[i + time_step, 0])
    return np.array(X), np.array(y)

# LSTM Model
def build_lstm_model(input_shape):
    model = Sequential()
    model.add(LSTM(50, return_sequences=True, input_shape=input_shape))
    model.add(LSTM(50, return_sequences=False))
    model.add(Dense(25))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Predict stock prices and compute metrics
def predict_and_evaluate_stock(stock_name):
    df_stock, scaled_data, scaler = preprocess_data(df, stock_name)
    X, y = create_dataset(scaled_data)
    X = np.reshape(X, (X.shape[0], X.shape[1], 1))
    
    model = build_lstm_model((X.shape[1], 1))
    model.fit(X, y, batch_size=1, epochs=1)
    
    predictions = model.predict(X)
    predictions = scaler.inverse_transform(predictions)
    actuals = scaler.inverse_transform(y.reshape(-1, 1))
    
    # Save results
    df_stock['Predicted Close'] = np.nan
    df_stock.iloc[60:, df_stock.columns.get_loc('Predicted Close')] = predictions[:, 0]
    
    # Compute performance metrics (binary classification for up or down)
    y_pred_binary = (predictions[1:] > actuals[:-1]).astype(int)
    y_true_binary = (actuals[1:] > actuals[:-1]).astype(int)
    
    accuracy = accuracy_score(y_true_binary, y_pred_binary)
    recall = recall_score(y_true_binary, y_pred_binary)
    precision = precision_score(y_true_binary, y_pred_binary)
    f1 = f1_score(y_true_binary, y_pred_binary)
    tn, fp, fn, tp = confusion_matrix(y_true_binary, y_pred_binary).ravel()
    specificity = tn / (tn + fp)
    
    metrics = {
        'Accuracy': accuracy,
        'Recall (Sensitivity)': recall,
        'Precision': precision,
        'F1 Score': f1,
        'Specificity': specificity
    }
    
    return df_stock, metrics

# Visualize the results
def visualize_results(stock_name, df_stock):
    plt.figure(figsize=(16,8))
    plt.title(f'{stock_name} Stock Price Prediction')
    plt.plot(df_stock['date'], df_stock['close'], label='Actual Close Price', color='blue')
    plt.plot(df_stock['date'], df_stock['Predicted Close'], label='Predicted Close Price', color='red')
    plt.xlabel('Date')
    plt.ylabel('Close Price USD ($)')
    plt.legend()
    plt.show()

# Main Loop: Process all stocks and save to one Excel file with multiple sheets
with pd.ExcelWriter(united_csv_path, engine='xlsxwriter') as writer:
    all_metrics = {}

    for stock_name in df['stock'].unique():
        print(f"Processing stock: {stock_name}")
        df_stock, metrics = predict_and_evaluate_stock(stock_name)
        visualize_results(stock_name, df_stock)
        
        # Add prediction results to a new sheet in the Excel file
        df_stock.to_excel(writer, sheet_name=f'{stock_name}_predictions', index=False)
        
        # Collect metrics
        all_metrics[stock_name] = metrics
        
        # Display the metrics table for this stock
        print(pd.DataFrame(metrics, index=[0]))

# Save all metrics to a CSV file
metrics_df = pd.DataFrame(all_metrics).T
metrics_df.to_excel(writer, sheet_name='all_stock_metrics')

# Display the metrics for the first few stocks
print(metrics_df.head())
```

* **The Transformer model!**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix
from keras.models import Model, Sequential
from keras.layers import Dense, Input, LSTM, RepeatVector, TimeDistributed, Flatten, BatchNormalization
from keras.layers import MultiHeadAttention, LayerNormalization, Add, Dropout, Reshape
from keras.optimizers import Adam
from keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
import os

# File directories
data_path = r"C:\Users\Jaber\OneDrive - University of Florida\Educational\GitHub\StockMarket\Data\all_stocks_5yr.csv"
results_dir = r"C:\Users\Jaber\OneDrive - University of Florida\Educational\GitHub\StockMarket\Results"
united_csv_path = os.path.join(results_dir, 'all_stocks_predictions_metrics.xlsx')

# Create results directory if it doesn't exist
if not os.path.exists(results_dir):
    os.makedirs(results_dir)

# Load dataset
df = pd.read_csv(data_path)

# Data Preprocessing
def preprocess_data(df, stock_name):
    df_stock = df[df['stock'] == stock_name].copy()
    df_stock.sort_values('date', inplace=True)
    
    # Scale data
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(df_stock[['close']].values)
    
    return df_stock, scaled_data, scaler

# Create the dataset for the model with dynamic window size
def create_dataset(scaled_data, max_time_step=200):
    X, y = [], []
    time_step = min(max_time_step, len(scaled_data) - 1)  # Adjust window size if data is insufficient
    for i in range(len(scaled_data) - time_step):
        X.append(scaled_data[i:i + time_step, 0])
        y.append(scaled_data[i + time_step, 0])
    return np.array(X), np.array(y)

# Transformer Layer
def transformer_encoder(inputs, head_size, num_heads, ff_dim, dropout=0):
    # Normalization and Attention
    x = LayerNormalization(epsilon=1e-6)(inputs)
    x = MultiHeadAttention(key_dim=head_size, num_heads=num_heads, dropout=dropout)(x, x)
    x = Dropout(dropout)(x)
    res = Add()([x, inputs])

    # Feed Forward Part
    x = LayerNormalization(epsilon=1e-6)(res)
    x = Dense(ff_dim, activation="relu")(x)
    x = Dropout(dropout)(x)
    x = Dense(inputs.shape[-1])(x)
    return Add()([x, res])

# Transformer Model without Autoencoder
def build_transformer_model(input_shape, head_size=256, num_heads=4, ff_dim=128, num_transformer_blocks=4, mlp_units=[128], dropout=0.2):
    inputs = Input(shape=(input_shape[1], input_shape[2]))
    
    # Transformer Layers
    x = inputs
    for _ in range(num_transformer_blocks):
        x = transformer_encoder(x, head_size, num_heads, ff_dim, dropout)
    
    # Flatten and MLP Head
    x = Flatten()(x)
    for units in mlp_units:
        x = Dense(units, activation="relu")(x)
        x = Dropout(dropout)(x)
    x = Dense(1)(x)
    model = Model(inputs, x)
    model.compile(optimizer=Adam(learning_rate=1e-4), loss="mse")
    return model

# Predict stock prices and compute metrics
def predict_and_evaluate_stock(stock_name):
    df_stock, scaled_data, scaler = preprocess_data(df, stock_name)
    X, y = create_dataset(scaled_data)
    X = np.reshape(X, (X.shape[0], X.shape[1], 1))
    
    model = build_transformer_model((X.shape[0], X.shape[1], 1))
    
    # Early stopping to prevent overfitting
    early_stopping = EarlyStopping(monitor='loss', patience=5, restore_best_weights=True)
    
    model.fit(X, y, batch_size=32, epochs=50, verbose=1, callbacks=[early_stopping])
    
    predictions = model.predict(X)
    predictions = scaler.inverse_transform(predictions)
    actuals = scaler.inverse_transform(y.reshape(-1, 1))
    
    # Save results
    df_stock['Predicted Close'] = np.nan
    df_stock.iloc[len(scaled_data) - len(predictions):, df_stock.columns.get_loc('Predicted Close')] = predictions[:, 0]
    
    # Compute performance metrics (binary classification for up or down)
    y_pred_binary = (predictions[1:] > actuals[:-1]).astype(int)
    y_true_binary = (actuals[1:] > actuals[:-1]).astype(int)
    
    accuracy = accuracy_score(y_true_binary, y_pred_binary)
    recall = recall_score(y_true_binary, y_pred_binary)
    precision = precision_score(y_true_binary, y_pred_binary)
    f1 = f1_score(y_true_binary, y_pred_binary)
    tn, fp, fn, tp = confusion_matrix(y_true_binary, y_pred_binary).ravel()
    specificity = tn / (tn + fp)
    
    metrics = {
        'Accuracy': accuracy,
        'Recall (Sensitivity)': recall,
        'Precision': precision,
        'F1 Score': f1,
        'Specificity': specificity
    }
    
    return df_stock, metrics

# Visualize the results
def visualize_results(stock_name, df_stock):
    plt.figure(figsize=(16,8))
    plt.title(f'{stock_name} Stock Price Prediction')
    plt.plot(df_stock['date'], df_stock['close'], label='Actual Close Price', color='blue')
    plt.plot(df_stock['date'], df_stock['Predicted Close'], label='Predicted Close Price', color='red')
    plt.xlabel('Date')
    plt.ylabel('Close Price USD ($)')
    plt.legend()
    plt.show()

# Main Loop: Process all stocks and save to one Excel file with multiple sheets
with pd.ExcelWriter(united_csv_path, engine='xlsxwriter') as writer:
    all_metrics = {}

    for stock_name in df['stock'].unique():
        print(f"Processing stock: {stock_name}")
        df_stock, metrics = predict_and_evaluate_stock(stock_name)
        visualize_results(stock_name, df_stock)
        
        # Add prediction results to a new sheet in the Excel file
        df_stock.to_excel(writer, sheet_name=f'{stock_name}_predictions', index=False)
        
        # Collect metrics
        all_metrics[stock_name] = metrics
        
        # Display the metrics table for this stock
        print(pd.DataFrame(metrics, index=[0]))

# Save all metrics to a CSV file
metrics_df = pd.DataFrame(all_metrics).T
metrics_df.to_excel(writer, sheet_name='all_stock_metrics')

# Display the metrics for the first few stocks
print(metrics_df.head())
```
